```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names= ['Estacion del año',
'Edad',
'Enfermedades infantiles',
'accidente o trauma grave',
'Intervencion qx', 'Fiebres altas en el ultimo año', 'Frecuencia en consumo de alcohol', 'Habito de fumar',
        'Numero de horas dedicadas a sentarse por dia', 'Salida']
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00244/fertility_Diagnosis.txt', names=names)
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
x
```

```python
y = pd.get_dummies(y)
```

```python
y
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model=MLPRegressor(hidden_layer_sizes=(100,100,20),max_iter=6000000)
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict(x)
```

```python
plt.plot(model.predict(x), 'ob')
plt.plot(y,'or')
```
